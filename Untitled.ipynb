```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import re
```

```python
recipies = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DS0103EN/labs/data/recipes.csv')
```

```python
recipies.head()
```

```python
recipies.shape
```

```python
items= list(recipies.columns.values)
items
```

```python
#types of searches using Re, match.group takes the second for loop and looks for 'rice'
print([match.group(0) for item in items for match in [(re.compile(".*(rice).*")).search(item)]if match])
print([match.group(0) for item in items for match in [(re.compile(".*(wasabi).*")).search(item)]if match])
print([match.group(0) for item in items for match in [(re.compile(".*(soy).*")).search(item)]if match])
```
